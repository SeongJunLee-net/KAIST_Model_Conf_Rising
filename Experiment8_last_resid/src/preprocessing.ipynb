```python
import json
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
all = pd.read_json('../dataset/all.json')
all
```

```python
# split data by using stratify for rel_id
X_train, X_test, y_train, y_test = train_test_split(all,all.loc[:,'rel_id'].values,
                                                    test_size=0.15, random_state=42, stratify=all['rel_id'])
```

```python
X_train.reset_index(inplace=True,drop=True)
X_test.reset_index(inplace=True,drop=True)
```
